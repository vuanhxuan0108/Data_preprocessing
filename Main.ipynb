##### a. Xét cột city

##### ---- Dùng phương pháp Frequency Encoding cho biến "city" do có số lương lớn giá trị, thay thế bằng tần xuất xuất hiện
##### ---- Frequency Encoding: Sử dụng khi biến phân loại có số lượng giá trị lớn và bạn muốn giữ lại thông tin về tần suất xuất hiện của mỗi giá trị.

```python
# In ra tần suất xuất hiện của các giá trị phân biệt trong cột 'city'
df['city'].value_counts()
```

```python
# Tính xác suất xuất hiện của các giá trị phân biệt, chuyển thành từ điển
frequency_count = (df['city'].value_counts() / len(df['city']) ).to_dict()
print(frequency_count)
```

```python
# hàm map có nhiệm vụ ánh xạ giữa key sang value, gán lại cho cột city
df['city'] = df['city'].map(frequency_count)
```

```python
# Kiểu dữ liệu cột city sau khi xử lý
df['city'].dtype
```

```python
# In ra dữ liệu sau khi mã hóa cột 'city'
df.head()
```

##### ---- Dùng Ordinal Encoding cho các cột relevent_experience, enrolled_university, education_level, company_size, company_type
##### ---- Ordinal Encoding: Sử dụng khi biến phân loại có thứ tự và các giá trị có mối quan hệ với nhau.

```python
# In ra các giá trị phân biệt các cột dữ liệu
print(df['enrolled_university'].unique())
print(df['education_level'].unique())
print(df['company_size'].unique())
print(df['company_type'].unique())
print(df['relevent_experience'].unique())
```

```python
# - Thực hiện tạo các từ điển và gán các value có thứ tự cho các giá trị tăng dần (đảm bảo luôn bắt đầu từ 0)
label_enrolled_university = {'no_enrollment':0, 'Full time course':2, 'Part time course':1, 'missing':3}
label_education_level = {'Graduate':2, 'Masters':3, 'High School':1, 'Phd':4, 'Primary School':0, 'missing':5}
label_company_size = {'50-99':2, '<10':0, '10000+':7, '5000-9999':6, '1000-4999':5, '10-49':1, '100-500':3, '500-999':4, 'missing':8}
label_company_type = {'Pvt Ltd':2, 'Funded Startup':3, 'Early Stage Startup':4, 'Other':5, 'Public Sector':0,'NGO':1, 'missing':6}
label_relevent_experience = {'Has relevent experience':1, 'No relevent experience':0}
```

```python
# Thực hiện dùng hàm map để ánh xạ các cặp key - value tương ứng
df['enrolled_university'] = df['enrolled_university'].map(label_enrolled_university)
df['education_level'] = df['education_level'].map(label_education_level)
df['company_size'] = df['company_size'].map(label_company_size)
df['company_type'] = df['company_type'].map(label_company_type)
df['relevent_experience'] = df['relevent_experience'].map(label_relevent_experience)
```

```python
# Vẽ biểu đồ boxplot để xem phân phối dữ liệu và phát hiện phần tử ngoại lệ
sns.boxplot( y='city_development_index', data=df)
```

```python
IQR = df["city_development_index"].quantile(0.75) - df["city_development_index"].quantile(0.25)
lower_city_limit = df["city_development_index"].quantile(0.25) - (IQR * 1.5)
upper_city_limit = df["city_development_index"].quantile(0.75) + (IQR * 1.5)

print(lower_city_limit)
print(upper_city_limit)
```

```python
# Những giá trị lớn hơn cận trên hoặc nhỏ hơn cận dưới gán True, còn lại gán False
outlier_city_development_index= np.where(df["city_development_index"] > upper_city_limit, True,
            np.where(df["city_development_index"] < lower_city_limit, True, False))
```

```python
# Lọc theo điều kiện
df = df.loc[~outlier_city_development_index]
```

```python
# Những giá trị lớn hơn cận gán bằng cận trên, nhỏ hơn cận dưới gán bằng cận dưới, còn lại giữ nguyên
df["training_hours"] = np.where(df['training_hours'] > upper_hours_limit, upper_hours_limit,
             np.where(df["training_hours"] < lower_hours_limit, lower_hours_limit, df["training_hours"]))
```

# II, Ứng dụng mô hình học máy cho dữ liệu bài toán

```python
# Khai báo thư viện cần thiết để phân chia dữ liệu
from sklearn.model_selection import train_test_split
```

```python
# Dữ liệu đầu vào
df.head(10)
```

```python
# Kích thước dữ liệu đầu vào
df.shape
```

```python
# Xóa cột enrollee_id, không làm ảnh hưởng đến mô hình
df.drop(['enrollee_id'], axis=1, inplace=True)
```

```python
# In lại dữ liệu
df.head()
```

```python
# Chia dữ liệu thành tập train:test 70:30 cột target là nhãn, các cột còn lại là đặc trưng
dtTrain, dtTest = train_test_split(df, test_size=0.2, shuffle=False)
y_train = dtTrain.loc[:, 'target']
y_test = dtTest.loc[:, 'target']
X_train = dtTrain.drop(['target'], axis=1)
X_test = dtTest.drop(['target'], axis=1)
```

#### Dùng mô hình Support Vector Meachine

```python
# Khai báo thư viện mô hình SVM
from sklearn.svm import SVC
```

```python
# Huấn luyện mô hình với tập dữ liệu trên
model = SVC(C=10)
model.fit(X_train.values, y_train.values)
y_pred = model.predict(X_test)
```

```python
# In ra các độ đo đánh giá mô hình
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Ti le du doan dung theo SVM:")
print('accuracy_score:', accuracy_score(y_test,y_pred))
print('precision_score:', precision_score(y_test,y_pred))
print('recall_score:', recall_score(y_test,y_pred))
print('f1_score:', f1_score(y_test,y_pred))
```

```python
# Biểu đồ thể hiện độ chính xác của mô hình qua các chỉ số dự đoán

import matplotlib.pyplot as plt
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
metrics_values = [accuracy_score(y_test, y_pred),
                  precision_score(y_test, y_pred),
                  recall_score(y_test, y_pred),
                  f1_score(y_test, y_pred)]

plt.bar(metrics_names, metrics_values, color=['blue', 'green', 'red', 'purple'])
plt.title('SVM Model Evaluation')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.show()
```
