# Part 2

##### Từ điển trong Python

```python
person = {
    "first" : "Corey",
    "last" : "Schafer", 
    "email" : "CoreyMSchafer@gmail.com"
}
```

```python
people = {
    "first" : ["Corey"],
    "last" : ["Schafer"], 
    "email" : ["CoreyMSchafer@gmail.com"]
}
```

```python
people = {
    "first" : ["Corey", "Jane", "John"],
    "last" : ["Schafer", "Doe", "Doe"], 
    "email" : ["CoreyMSchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com"]
}
```

```python
people["email"]
```

##### Tạo dataframe từ từ điển mới xây dựng bên trên

```python
df = pd.DataFrame(people)
```

```python
df
```

##### Mỗi cột của từ điển là một series

```python
type(df.email)
```

##### Cách truy vấn ra n cột bất kỳ khi biết tên cột

```python
df[['email', 'last']]
```

##### Cách truy vấn cột và hàng mong muốn

```python
df[1:][['first', 'last']]
```

##### Đưa ra tên các thuộc tính trong DataFrame

```python
df.columns
```

##### Cách dùng loc và iloc

##### >>>>> iloc dùng để truy cập hàng, côt theo vị trí số nguyên
##### df.iloc[hàng, cột]

##### Đưa ra mẫu số 0 của dữ liệu

```python
df.iloc[0]
```

##### Đưa ra mẫu số 0 và 2 của dữ liệu

```python
df.iloc[[0, 2]]
```

##### Đưa ra cột 0 và 2 (first và email)

```python
df.iloc[:, [0, 2]]
```

##### Đưa ra mẫu số 0 và 2 nhưng chỉ hiển thị cột 2 (email)

```python
df.iloc[[0, 2], 2]
```

##### >>>>> loc dùng để truy cập hàng, côt theo nhãn

##### Đưa ra mẫu số 0 củ dữ liệu

```python
df.loc[0]
```

##### Đưa ra mẫu số 0 và 2 của dữ liệu

```python
df.loc[[0, 2]]
```

##### Đưa ra mẫu sô 0 và 2 của dữ liệu nhưng chỉ hiển thị cột email

```python
df.loc[[0, 2], "email"]
```

##### Đưa ra cột 1 và 2 (last và email)

```python
df.loc[:, 'last':]
```
