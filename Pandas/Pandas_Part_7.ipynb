# Hướng dẫn Python Pandas (Phần 7): Sắp xếp dữ liệu

## Part 1

##### Khai báo thư viện

```python
import pandas as pd 
```

##### Tạo từ điển people

```python
people = {
    "first" : ["Corey", "Jane", "John", "Adam"],
    "last" : ["Schafer", "Doe", "Doe", "Doe"], 
    "email" : ["CoreyMSchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com", "AdamDoe@gmail.com"]
}
```

##### Tạo dataframe từ từ điển trên

```python
df = pd.DataFrame(people)
```

##### In ra dữ liệu

```python
df
```

##### Sắp xếp giá trị cột 'last' tăng dần

```python
df.sort_values(by='last', ascending=True)
## ascending = True là tăng dần, ascending = Fasle là giảm dần
```

##### Sắp xếp cột first và last tăng dần

```python
df.sort_values(by=['last', 'first'], ascending=True)
```

##### Sắp xếp cột 'first' tăng dần và cột 'last' giảm dần

```python
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df
```

##### Sắp xếp lại chỉ mục

```python
df.sort_index(axis=0)
## axis = 0 là sắp xếp lại chỉ mục hàng, axis = 1 là sắp xếp lại chỉ mục cột
```

##### Sắp xếp riêng cột 'last'

```python
df['last'].sort_values()
```
