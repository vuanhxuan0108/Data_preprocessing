##### Đưa ra phần trăm tỉ lệ sử dụng mạng xã hội của các quốc gia

```python
grp_Country = df['SocialMedia'].groupby(df['Country']).value_counts(normalize=True)
```

##### Gom nhóm dữ liệu theo cột 'Country'

```python
grp_Country = df.groupby(df['Country'])
```

##### In ra dữ liệu 'Viet Nam', lấy ra cột 'Hobbyist' (2 cách)

```python
grp_Country.get_group('Viet Nam')[['Hobbyist']]
```
