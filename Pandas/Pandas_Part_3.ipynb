# Hướng dẫn Python Pandas (Phần 3): Chỉ mục - Cách đặt, đặt lại và sử dụng chỉ mục

##### Khai báo thư viện pandas

```python
import pandas as pd
```

# Part 1

##### Khởi tạo từ điển people

```python
people = {
    "first" : ["Corey", "Jane", "John"],
    "last" : ["Schafer", "Doe", "Doe"], 
    "email" : ["CoreyMSchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com"]
}
```

##### Tạo dataframe từ từ điển people

```python
df = pd.DataFrame(people)
```

##### Hiển thị dữ liệu đầu vào

```python
df
```

##### In ra chỉ mục

```python
df.index
```

##### Lấy ra cột 2 (email)

```python
df['email']
```

##### Cách đặt lại chỉ mục

```python
df.set_index('email', inplace=True)
```

##### Dữ liệu với chỉ mục là cột email

```python
df
```

##### In ra chỉ mục

```python
df.index
```

##### In ra dữ liệu từ hàng thứ 1 trở đi, chỉ gồm cột last

```python
df.loc['JaneDoe@gmail.com':, 'last']
```

##### In ra dữ liệu từ hàng thứ 1 trở đi, chỉ gồm cột last

```python
df.iloc[1:, 1]
```

##### Cách reset chỉ mục về ban đầu

```python
df.reset_index(inplace=True)
```

##### In ra dữ liệu sau khi reset

```python
df
```

##### Đặt thuộc tính đầu tiên làm chỉ mục

```python
df.set_index('Respondent', inplace=True)
```

##### In ra 5 dòng đầu tiên của dữ liệu

```python
df.head()
```

##### Chỉ mục thường đặt là cột mà các giá trị chỉ xuất hiện 1 lần duy nhất, để có thể dễ dàng truy xuất

##### Đặt cột đầu tiên của dữ liệu mô tả để tiện cho việc xem mô tả của các thuộc tính

```python
schedma_df.set_index('Column', inplace = True)
```

##### In ra dữ liệu sau khi đặt lại chỉ mục

```python
schedma_df
```

##### Lấy ra mô tả của 'IDE' trong dữ liệu mô tả

```python
schedma_df.loc['IDE', 'QuestionText']
```

##### Sắp xếp dữ liệu theo chữ cái theo cột QuestionText

```python
schedma_df.sort_values('QuestionText', ascending=False, inplace=True)
# ascending = Fasle thì sắp xếp giảm dần, ascending = True thì sắp xếp tăng dần
# inplace = True thì thay đổi thực sự, inplace = Fasle là trải nghiệm tức thời
```

##### Dự liệu mô tả sau khi sắp xếp

```python
schedma_df
```

##### Sắp xếp dữ liệu theo chỉ mục

```python
schedma_df.sort_index(ascending=True, inplace=True)
# ascending = Fasle thì sắp xếp giảm dần, ascending = True thì sắp xếp tăng dần
# inplace = True thì thay đổi thực sự, inplace = Fasle là trải nghiệm tức thời
```

##### Dự liệu mô tả sau khi sắp xếp

```python
schedma_df
```
